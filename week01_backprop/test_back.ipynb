```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
np.random.shuffle(x)
y = [x[i] * i for i in range(len(x))] 
y = np.array(y) + 5
W = np.random.sample(x.shape)
b = np.random.sample(1)
print(x)
print(y)
print(W)
print(b)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, x*W + b)
```

```python
np.sum(np.array([1, 2]), axis=0)
```

```python
def MSE(predicted, actual):
    return np.sum((actual - predicted)**2, axis=0) / len(predicted) 
```

```python
predict = x * W + b
loss = MSE(predict, y)
loss
```

```python
delta = 0.00001
for epoch in range(100):
    predict = x * W + b
    loss = MSE(predict, y)
    print(f"Loss => {loss}")
    W = W - delta * 2*(y - predict)*(-x)
    b = b - delta * 2*(y - predict)
```
